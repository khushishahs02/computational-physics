## Energy Conservation in the Euler-Cromer Method

We address two goals:

1. **Derive** the energy difference $E_{n+1} - E_n$ for the Euler-Cromer algorithm applied to $\ddot{x} = -x$.
2. **Show** that for simple harmonic motion, the leading error terms average to zero over one complete cycle — so Euler-Cromer conserves energy up to $\mathcal{O}(\Delta t^4)$.

The mechanical energy at time step $n$ is:

$$
E_n = \frac{1}{2}\left(v_n^2 + x_n^2\right)
$$


---
## 🔹 Step 0: What We Are Trying to Prove

The total energy of the simple harmonic oscillator $\ddot{x} = -x$ is:

$$
E = \frac{1}{2}\left(v^2 + x^2\right)
$$

For an **exact** solution, this energy is constant at every instant of time. Any numerical method will introduce errors, so the question is: *how badly does the energy drift, and how fast?*

We want to show two things for the Euler-Cromer algorithm:

- The **per-step** energy change $E_{n+1} - E_n$ is small — specifically, it starts at order $(\Delta t)^2$, with **no** $(\Delta t)^1$ term.
- The lower-order error terms ($(\Delta t)^2$ and $(\Delta t)^3$) **average to exactly zero** over one complete oscillation cycle for SHM, leaving only a $(\Delta t)^4$ residual.

This is what makes Euler-Cromer far superior to the standard Euler method for oscillatory problems.


---
## 🔹 Step 1: Write the Euler-Cromer Update Rules

For the oscillator $\ddot{x} = -x$, the Euler-Cromer algorithm updates velocity first, then uses the **new** velocity to update position:

$$
v_{n+1} = v_n - \Delta t\, x_n \tag{1}
$$

$$
x_{n+1} = x_n + \Delta t\, v_{n+1} \tag{2}
$$

> **Why this matters:** In the standard Euler method, equation (2) would use the *old* $v_n$ instead of $v_{n+1}$. That single change — using the already-updated velocity — is responsible for all the improved energy behaviour we derive below.


---
## 🔹 Step 2: Write the Energy at the Next Step

We want to compute:

$$
E_{n+1} = \frac{1}{2}\left(v_{n+1}^2 + x_{n+1}^2\right)
$$

To find $E_{n+1} - E_n$, we substitute the update rules (1) and (2) and expand everything in powers of $\Delta t$, keeping all terms up to $(\Delta t)^4$. We do this in two parts — the velocity term and the position term — then add them.


---
## 🔹 Step 3: Expand the Velocity Term $v_{n+1}^2$

From equation (1):

$$
v_{n+1} = v_n - \Delta t\, x_n
$$

Squaring:

$$
v_{n+1}^2 = \left(v_n - \Delta t\, x_n\right)^2
$$

$$
\boxed{v_{n+1}^2 = v_n^2 - 2\,\Delta t\, v_n x_n + (\Delta t)^2 x_n^2} \tag{3}
$$


---
## 🔹 Step 4: Expand the Position Term $x_{n+1}^2$

Substitute equation (1) into equation (2):

$$
x_{n+1} = x_n + \Delta t\,(v_n - \Delta t\, x_n)
$$

$$
x_{n+1} = x_n + \Delta t\, v_n - (\Delta t)^2 x_n \tag{4}
$$

Now square this expression, expanding all cross-terms and keeping everything up to $(\Delta t)^4$:

$$
x_{n+1}^2 = \left(x_n + \Delta t\, v_n - (\Delta t)^2 x_n\right)^2
$$

$$
\boxed{x_{n+1}^2 = x_n^2 + 2\,\Delta t\, x_n v_n - 2(\Delta t)^2 x_n^2 + (\Delta t)^2 v_n^2 - 2(\Delta t)^3 v_n x_n + (\Delta t)^4 x_n^2} \tag{5}
$$


---
## 🔹 Step 5: Add Both Terms

We compute $v_{n+1}^2 + x_{n+1}^2$ by adding equations (3) and (5):

**From velocity** (eq. 3):

$$
v_n^2 \;-\; 2\,\Delta t\,v_n x_n \;+\; (\Delta t)^2 x_n^2
$$

**From position** (eq. 5):

$$
x_n^2 \;+\; 2\,\Delta t\,x_n v_n \;-\; 2(\Delta t)^2 x_n^2 \;+\; (\Delta t)^2 v_n^2 \;-\; 2(\Delta t)^3 v_n x_n \;+\; (\Delta t)^4 x_n^2
$$

Total sum, grouped by power of $\Delta t$:

$$
= \underbrace{v_n^2 + x_n^2}_{\Delta t^0}
\underbrace{- 2\Delta t\,v_n x_n + 2\Delta t\,x_n v_n}_{\Delta t^1}
+ \underbrace{(\Delta t)^2(x_n^2 - 2x_n^2 + v_n^2)}_{\Delta t^2}
- \underbrace{2(\Delta t)^3 v_n x_n}_{\Delta t^3}
+ \underbrace{(\Delta t)^4 x_n^2}_{\Delta t^4}
$$


---
## 🔴 Step 6: Simplify — The Key Cancellation

### Cancellation of the $\Delta t$ term

$$
-2\,\Delta t\,v_n x_n + 2\,\Delta t\,x_n v_n = 0
$$

> **This is the most important step.** The first-order term vanishes completely.
> In the standard Euler method (which uses old $v_n$ in the position update), > these two terms are **not equal** and do not cancel, leaving a residual > $\mathcal{O}(\Delta t)$ energy gain at every step. > Euler-Cromer eliminates this by construction.

### Simplify the $\Delta t^2$ term

$$
(\Delta t)^2(x_n^2 - 2x_n^2 + v_n^2) = (\Delta t)^2(v_n^2 - x_n^2)
$$

### Full simplified sum

$$
v_{n+1}^2 + x_{n+1}^2 = v_n^2 + x_n^2 + (\Delta t)^2(v_n^2 - x_n^2) - 2(\Delta t)^3 v_n x_n + (\Delta t)^4 x_n^2
$$


---
## 🔹 Step 7: Subtract the Original Energy

Since $E_n = \tfrac{1}{2}(v_n^2 + x_n^2)$, dividing by 2 and subtracting $E_n$:

$$
\boxed{
E_{n+1} - E_n = (v_n^2 - x_n^2)\,(\Delta t)^2 - 2\,v_n x_n\,(\Delta t)^3 + x_n^2\,(\Delta t)^4
}
$$

✅ **First part proved.**

The energy change per step has **no $\Delta t$ term** — the lowest-order error is $(\Delta t)^2$. But this still means energy drifts at every step. The question is whether this drift accumulates or cancels over a full oscillation cycle — which we now investigate using the physics of SHM.


---
## 🔹 Step 8: Why One Complete Cycle Matters

The energy change $E_{n+1} - E_n$ is non-zero at each individual time step. So why does Euler-Cromer still work so well for oscillatory problems?

The key insight is that we should not look at individual steps — we should ask how much **net energy change accumulates over one full oscillation cycle**.

For an oscillatory system like SHM:

- At some phases of the cycle the energy increases $(E_{n+1} > E_n)$.
- At other phases the energy decreases $(E_{n+1} < E_n)$.

If these gains and losses **exactly cancel** when summed over one period, the net drift per cycle is zero. This would mean the algorithm conserves energy *on average* — which is precisely the behaviour we observe in simulation.

To test this, we replace $x_n$ and $v_n$ with the **exact SHM solution** and average each coefficient over one period $T = 2\pi$:

$$
x(t) = \cos(t), \qquad v(t) = -\sin(t), \qquad T = 2\pi
$$

> **Why SHM specifically?** The oscillator $\ddot{x} = -x$ has $\omega_0 = 1$. > We use the exact analytical solution to evaluate the averages cleanly. > If the cancellation holds for the exact solution, it confirms that Euler-Cromer's > energy errors are genuinely oscillatory (not drifting) for SHM.


---
## 🔹 Step 9: Average Each Coefficient Over One Period

### 🔴 Term 1 — Coefficient of $(\Delta t)^2$: Average of $(v^2 - x^2)$

Substituting the SHM solution:

$$
v^2 - x^2 = \sin^2(t) - \cos^2(t) = -\cos(2t)
$$

This is a pure cosine at **twice the natural frequency**. Its average over one period:

$$
\langle v^2 - x^2 \rangle = \frac{1}{2\pi}\int_0^{2\pi}(-\cos 2t)\,dt = -\frac{1}{2\pi}\left[\frac{\sin 2t}{2}\right]_0^{2\pi} = 0
$$

$$
\boxed{\langle v^2 - x^2 \rangle = 0}
$$

> **Physical meaning:** The $(\Delta t)^2$ term causes energy to rise when > $|v| > |x|$ (near equilibrium, high speed) and fall when $|x| > |v|$ > (near turning points, low speed). Over one complete cycle both phases are > visited equally — the gains and losses balance exactly, leaving zero net drift. > Without the full cycle this cancellation would not be complete.


### 🔴 Term 2 — Coefficient of $(\Delta t)^3$: Average of $(-2vx)$

Substituting the SHM solution:

$$
-2vx = -2(-\sin t)(\cos t) = 2\sin(t)\cos(t) = \sin(2t)
$$

Again a pure sinusoid at twice the natural frequency. Its average:

$$
\langle -2vx \rangle = \frac{1}{2\pi}\int_0^{2\pi}\sin(2t)\,dt = \frac{1}{2\pi}\left[-\frac{\cos 2t}{2}\right]_0^{2\pi} = 0
$$

$$
\boxed{\langle -2vx \rangle = 0}
$$

> **Physical meaning:** The $(\Delta t)^3$ error term is proportional to $vx$, > which is positive in two quadrants (both positive or both negative) and negative > in the other two. Over a full cycle all four quadrants are visited, and the > positive and negative contributions cancel completely.


### 🔴 Term 3 — Coefficient of $(\Delta t)^4$: Average of $x^2$

Substituting the SHM solution:

$$
x^2 = \cos^2(t) = \frac{1 + \cos(2t)}{2}
$$

Its average:

$$
\langle x^2 \rangle = \frac{1}{2\pi}\int_0^{2\pi}\cos^2(t)\,dt = \frac{1}{2} \neq 0
$$

$$
\boxed{\langle x^2 \rangle = \frac{1}{2}}
$$

This term does **not** vanish. The residual energy drift per cycle is:

$$
\Delta E_{\text{cycle}} \approx \frac{1}{2}\cdot(\Delta t)^4 \cdot \underbrace{\frac{2\pi}{\Delta t}}_{N_{\text{steps}}} = \pi\,(\Delta t)^3
$$

> **Physical meaning:** $x^2 \geq 0$ always — it never goes negative. > So this term always makes a positive contribution to the energy. > It cannot cancel over a cycle the way the sinusoidal terms did. > However, because it is multiplied by $(\Delta t)^4$, its per-step contribution > is extremely small for small $\Delta t$.


---
## 🔹 Step 10: Final Conclusion — What Each Power of $\Delta t$ Means

| Term | Coefficient | Average over $T$ | Net effect |
|------|-------------|------------------|------------|
| $(\Delta t)^1$ | cancelled algebraically | — | **Zero by construction** |
| $(\Delta t)^2$ | $v^2 - x^2 = -\cos(2t)$ | $0$ ✓ | **Cancels over one cycle** |
| $(\Delta t)^3$ | $-2vx = \sin(2t)$ | $0$ ✓ | **Cancels over one cycle** |
| $(\Delta t)^4$ | $x^2 = \cos^2(t)$ | $\tfrac{1}{2} \neq 0$ | Residual drift $\sim \pi(\Delta t)^3$/cycle |

### Why the cancellations happen: the role of a complete cycle

The coefficients of $(\Delta t)^2$ and $(\Delta t)^3$ are both sinusoids at frequency $2\omega_0$ — they oscillate at exactly **twice** the natural frequency. A fundamental property of sinusoids is that their integral over any **integer number of complete periods is exactly zero**. Since we average over one full oscillation period $T = 2\pi/\omega_0$, which contains exactly one period of $\cos(2t)$ and $\sin(2t)$, both integrals are zero.

**This cancellation only works if we complete a full cycle.** Over half a cycle, or an arbitrary time interval, the errors do not cancel — Euler-Cromer's advantage is specifically a long-time, oscillatory-problem property.

### What does $\Delta t^2$ being zero actually mean?

If the $(\Delta t)^2$ coefficient did *not* average to zero, the energy drift per cycle would be $\sim C \cdot (\Delta t)^2 \cdot N_{\text{steps}} = C \cdot (\Delta t)^2 \cdot (T/\Delta t) = C \cdot T \cdot \Delta t$. This would be a drift of $\mathcal{O}(\Delta t)$ per unit time — the same bad behaviour as the standard Euler method. By averaging to zero, this $\mathcal{O}(\Delta t)$ drift is completely eliminated.

### What does $\Delta t^3$ being zero actually mean?

Similarly, if only the $(\Delta t)^2$ term cancelled, the next leading drift would be $\sim (\Delta t)^3 \cdot N_{\text{steps}} = (\Delta t)^2$ per unit time. The $(\Delta t)^3$ coefficient also averaging to zero eliminates this too, pushing the energy drift down to:

$$
\Delta E_{\text{per unit time}} \sim (\Delta t)^4 \cdot \frac{1}{\Delta t} = (\Delta t)^3
$$

This is why we say **Euler-Cromer conserves energy up to $(\Delta t)^4$**: both lower-order error terms vanish over a cycle, and only the $(\Delta t)^4$ residual contributes to long-term drift — making the method far more accurate than its formal first-order truncation error would suggest.

## ✅ Summary

**Part 1 — Algebraic derivation:**

By substituting the Euler-Cromer updates into $E_{n+1} = \tfrac{1}{2}(v_{n+1}^2 + x_{n+1}^2)$ and expanding, the $\mathcal{O}(\Delta t)$ terms cancel exactly due to using $v_{n+1}$ in the position update. The result is:

$$
E_{n+1} - E_n = (v_n^2 - x_n^2)(\Delta t)^2 - 2v_n x_n(\Delta t)^3 + x_n^2(\Delta t)^4
$$

**Part 2 — Averaging over one SHM cycle:**

With $x = \cos t$, $v = -\sin t$, $T = 2\pi$:

- $(\Delta t)^2$ coefficient: $v^2 - x^2 = -\cos(2t)$ → average $= 0$ — eliminates $\mathcal{O}(\Delta t)$ long-term drift.
- $(\Delta t)^3$ coefficient: $-2vx = \sin(2t)$ → average $= 0$ — eliminates $\mathcal{O}(\Delta t^2)$ long-term drift.
- $(\Delta t)^4$ coefficient: $x^2 = \cos^2 t$ → average $= \tfrac{1}{2} \neq 0$ — irreducible residual, giving drift $\sim \pi(\Delta t)^3$ per cycle.

Both leading error terms are sinusoids at $2\omega_0$ whose integrals vanish over a complete period. This is not a coincidence — it is a direct consequence of the symplectic (phase-space-area-preserving) structure of the Euler-Cromer method, which guarantees that energy errors remain bounded and oscillatory rather than accumulating as a monotonic drift.
